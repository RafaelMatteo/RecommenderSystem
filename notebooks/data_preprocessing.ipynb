```python
import os
import requests
import pandas as pd

# Clave de API y ruta de la carpeta donde se van a guardar las imágenes
API_KEY = '#####d7215941c03f0e6f7500e4#####'
static_folder = '../static/images/movies_tmdb'  

# Crear la carpeta '../static/images/movies_tmdb' si no existe
os.makedirs(static_folder, exist_ok=True)

# Leer el archivo CSV con los 'tmdbId'
csv_file = '../data/raw/ml-latest-small/links.csv'
df = pd.read_csv(csv_file)

df = df.dropna(subset=['tmdbId'])
df['tmdbId'] = df['tmdbId'].astype(int)

# Iterar sobre cada 'tmdbId' en el CSV
for index, row in df.iterrows():
    tmdb_id = int(row['tmdbId'])
    
    # Nombre de archivo y la ruta completa
    file_name = f'{tmdb_id}.jpg'
    file_path = os.path.join(static_folder, file_name)
    
    # Vericador de existencia de imagen
    if os.path.exists(file_path):
        print(f"Imagen {file_name} ya existe, saltando descarga.")
        continue  # Si la imagen existe, pasar a la siguiente
    
    # Obtener detalles de la película
    url = f"https://api.themoviedb.org/3/movie/{tmdb_id}?api_key={API_KEY}"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        poster_path = data.get('poster_path')  # Ruta de la imágen
        
        if poster_path:  # Si hay una imágen
            # URL de la imagen de tamaño w92
            poster_url = f"https://image.tmdb.org/t/p/w92{poster_path}"
            
            # Descargar la imagen y guardarla
            img_data = requests.get(poster_url).content
            with open(file_path, 'wb') as handler:
                handler.write(img_data)
            
            print(f"Imagen de la película tmdb_id {tmdb_id} guardada en {file_path}")
        else:
            print(f"No hay imagen para el tmdb_id {tmdb_id}")
    else:
        print(f"Error al obtener la imagen tmdb_id {tmdb_id}: {response.status_code}")
```
